#### Load Neccessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load FIFA data
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

### Box and Whiskers Chart

```python
plt.figure(figsize=(5,6), dpi=120)

plt.style.use('default')

# Boxplot for Overall ratings of players from top football clubs
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
manchester = fifa.loc[fifa.Club == 'Manchester City']['Overall']

labels = ['FC Barcelona', 'Real Madrid', 'Manchester City']

# Create boxplot with custom median line properties, patch_artist to allow box fill color, medianprops to customize median line.
bp = plt.boxplot([barcelona, madrid, manchester], tick_labels=labels, patch_artist=True, medianprops={'linewidth': 2})


for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    box.set(hatch = '/')

plt.title('Overall Ratings Distribution of Top Football Clubs')
plt.ylabel('FIFA Overall Rating')


plt.show()
```
