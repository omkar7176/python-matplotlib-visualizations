#### Load Neccessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load FIFA data
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

#### Pie Chart 1

```python
# left and right footed players count
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()['Preferred Foot']
Right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()['Preferred Foot']

lables = ['Left', 'Right']

plt.pie([left, Right], labels=['Left', 'Right'], autopct='%.2f%%')

plt.title('Preferred Foot among FIFA Players')

# plt.legend(title='Preferred Foot', loc='upper right', bbox_to_anchor=(1.3,1))
plt.legend()

plt.show()
```

### Pie Chart 2

```python
# Set the figure size and dpi
plt.figure(figsize=(8,5), dpi=100)

# Set the style of the plot to ggplot style for better aesthetics
plt.style.use('ggplot')

# Clean the Weight column by removing 'lbs' and converting to integer type 
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

# Count players in different weight categories 
light = fifa.loc[fifa.Weight < 125].count()['Weight']
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()['Weight']
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()['Weight']
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()['Weight']
heavy = fifa[fifa.Weight >= 200].count()['Weight']

# Set the size of pie chart
weights = [light,light_medium, medium, medium_heavy, heavy]

# Define labels and explode parameters for better visualization
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']

# Explode the slices for 'under 125' and 'over 200' categories
# Explodes are is used to offset a slice from the center of the pie chart
explode = (.4,.2,0,0,.4)

plt.title('Ditribution of Player Weights in FIFA Data (in lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')

plt.show()
```
