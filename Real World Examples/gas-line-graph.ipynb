### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Real World Exmaples

#### Line Graph

```python
# Load the gas_prices.csv dataset
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(10, 6))

# Add title to the graph
plt.title('Gas Prices Over Time (in USD)', fontdict={'fontweight': 'bold', 'fontsize': 14})

# Plot each country's gas prices over the years
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'r.-', label='Canada')
plt.plot(gas['Year'], gas['South Korea'], 'g.-', label='South Korea')
plt.plot(gas.Year, gas.France, 'c.-', label='France')

# Alternative Way - to Plot each country's gas prices over the years using a loop
# countries_look_up = ['France', 'Germany', 'Italy']
# for country in gas:
#     if country != 'Year':
#     # if country in countries_look_up: # Only plot selected countries
#         plt.plot(gas['Year'], gas[country], marker='.', label=country)


plt.xlabel('Year')
plt.ylabel('USD')


#print(gas.Year[::3])  # Debugging line to check the years being used for ticks

plt.xticks(gas.Year[::3])  # Show every 3rd year on x-axis

plt.legend()

plt.savefig('gas_prices_line_graph.png', dpi=300)  # Save the figure as a PNG file

plt.show()
```
