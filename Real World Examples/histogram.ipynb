### Load Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Histogram

```python
# Load the fifa_data.csv dataset
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

```python
# Create histogram of player skill levels
bins = [40,50,60,70,80,90,100]

plt.figure(figsize=(8,5))

# Create histogram with specified bin sizes and color
# bins means the intervals for grouping data into ranges as 40-50, 50-60, etc as x-axis
plt.hist(fifa.Overall, bins=bins, color="#0252A1")

# Customize x-ticks to match bin edges 
plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```
