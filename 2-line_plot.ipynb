### Load Neccessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Rsize your graph(Creating a figure with specific size and resolution. figsize=(width, height) in inches, dpi=dots per inch(pixels per inch), when saving probably should use 300 if possible)
plt.figure(figsize=(6,5), dpi=170)

## Line 1

# Keyword Argument Notation(Using longhand notation for color, marker, and line style as below)
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', markersize=10, markeredgecolor='blue', markerfacecolor='black', linestyle='--') # Plotting the line graph with various customizations

# Shorthand Notation(Use shorhand notation for color, marker, and line style as below)
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

# Line 2

# Select interval we want to plot points at
x2 = np.arange(0,4.5,0.5) # np.arange(start, stop, step)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of the graph as dot graph
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Adding title and labels with font customization with fontdict
plt.title("Our First Graph", fontdict={ 'fontname': 'Comic Sans MS', 'fontsize': 20})

#  Adding labels to x-axis and y-axis
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# X, Y axis ticksmarks (scale of your graph)
plt.xticks([0,1,2,3,4,]) 
# plt.yticks([0,2,4,6,8,10])

# legend() to show labels of different lines.(This shows the label)
plt.legend()

# Save the graph as a file using savefig() function.(dpi 300 is good when saving so graph has high resolution)
plt.savefig("mygraph.png", dpi=170)

# Show the graph
plt.show()
```
